Let's start off by looking at the concepts of population and sample with an example. With a quick Google search, one can find a multitude of freely available datasets. The one used here collates details on nutritional value of different food types. It can be found <a href=https://graphia.app/example-data.html>here</a>.

This is a simple <i>.csv</i> file. In order to work with it, we'll simply download it to the same local directory as this notebook. Now we need to import it in the Jupyter environment. We can do that with <a href=https://www.w3schools.com/python/pandas/pandas_intro.asp>Pandas</a>, a very popular package for manipulating and analysing data. Here is how we import it to Python:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

##### Import dataset in dataframe format:

```python
path = "Food_contents_2019.csv"
data = pd.read_csv(path)
data.head() # see only the first 5 rows
```

##### Investigate what food groups are present in the dataframe

```python
data['Group'].unique()
```

##### Extract data concerning specific food products

```python
meat_df = data[data['Group']=='Meat and meat products ']
meat_df['Food Name'].unique()

alcohol_df = data[data['Group']=='Alcoholic beverages ']
alcohol_df['Food Name'].unique()

veg_df = data[data['Group']=='Vegetables']
veg_df['Food Name'].unique()

fish_df = data[data['Group']=='Fish and fish products ']
fish_df['Food Name'].unique()

fruit_df = data[data['Group']=='Fruit']
fruit_df['Food Name'].unique()

nuts_df = data[data['Group']=='Nuts and seeds ']
nuts_df['Food Name'].unique()
```

##### Enter nutritional value of interest
* In this case we will compare the protein content between different food groups

```python
value = 'Protein (g)'
```

###### To visualise the differences between samples and the whole population,we plot a histogram with respect to the nutritional value of interest.For instance the amount of protein. For convenient visualisation the package <a href=https://matplotlib.org/>matplotlib</a> is used:
